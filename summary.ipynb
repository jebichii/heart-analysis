## üåç The Problem  

Cardiovascular diseases are responsible for an estimated **17.9 million deaths annually**. Many of these are preventable through early diagnosis and affordable screening. Yet, in numerous low- and middle-income regions, access to accurate diagnostic tools is limited‚Äîhindering timely treatment. This challenge is directly targeted by **SDG 3** and **SDG 10: Reduced Inequalities**.

---

## üí° The Solution: Intelligent Screening with Machine Learning

Using a publicly available dataset (`heart.csv`), I developed an end-to-end machine learning pipeline designed for scalable health screening:

- **Data Preprocessing**: Cleaned and encoded clinical indicators such as age, chest pain type, blood pressure, cholesterol levels, etc.
- **Model Training**: Applied a robust **Random Forest Classifier** to learn predictive patterns
- **Evaluation**: Achieved strong metrics including high accuracy, F1-score, and low mean absolute error
- **Interpretability**: Visualized key risk factors (e.g., ST_Slope, ExerciseAngina) and confusion matrix for transparent model reporting

This pipeline serves as a lightweight and portable **diagnostic assistant**‚Äîparticularly valuable in regions where expert cardiologists may not be readily accessible.

## üîç Key Concepts Utilized

- **Supervised Learning**  
  Labeled heart data trains classification models to distinguish between patients with and without heart disease.

- **Data Preprocessing**
  - Managed missing values  
  - Scaled numerical features using `StandardScaler`  
  - Encoded categorical variables with `LabelEncoder`

- **Exploratory Data Analysis (EDA)**
  - Used visualizations (histograms, boxplots, heatmaps) to explore feature distributions and relationships

- **Model Selection & Optimization**
  - Explored Logistic Regression, Random Forest, and XGBoost  
  - Used Grid Search for hyperparameter optimization

- **Evaluation Metrics**
  - Accuracy  
  - Precision  
  - Recall  
  - F1-score  
  - Mean Absolute Error (MAE)
